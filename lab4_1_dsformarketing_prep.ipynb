```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/saleorder.csv", encoding='tis-620')
```

```python
df['s_date'] = pd.to_datetime(df['s_date'], format='%d %m %Y', errors='coerce') #เปลี่ยนให้เป็น datetime
```

```python
df.dtypes
```

```python
df['Month'] = pd.DatetimeIndex(df['s_date']).month #เพิ่ม column month
```

```python
df
```

```python
df_select = df[['Month','nettotal']] #เลือกข้อมูล month, nettotal
df_select #ใส่ไว้ใน df_select
```

```python
dp = pd.pivot_table(df_select, index=["Month"],values=["nettotal"],aggfunc=np.sum) #รวม index ซ้ำกันมารวมกันแล้วพลิกแนวนอนให้เป็นแนวตั้ง
```
